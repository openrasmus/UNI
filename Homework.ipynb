```
#Task 1:
#Perceptron to classify AND , OR
import matplotlib.pyplot as plt
import numpy as np


#"Neural network" implementation of AND with 2 inputs, x and y
#Truth table:
#0,0 -> return(-1>0) = False
#1,0 or 0,1 -> return(0>0) = False
#1,1 -> return(1>0 True)
def nn_and(x,y):
  b = -1
  w = 1
  return(w*x+w*y+b>0)

#Implementation of OR. 
def nn_or(x,y):
  b = 0
  w = 1
  return(w*x+w*y+b>0)

#Implementation of not, to make XOR:
def nn_not(x):
  b = 1
  w = -2
  return(x*w + b >0)

#Implementation of XOR
#https://medium.com/@lucaspereira0612/solving-xor-with-a-single-perceptron-34539f395182
#Xor can also be implemented with a single perceptron if you use transformations.
def nn_xor(x,y):
  return(nn_and(nn_not(nn_and(x,y)),nn_or(x,y)))


#uncomment to check that nn_xor is implemented correctly
#print(nn_xor(0,0))
#print(nn_xor(0,1))
#print(nn_xor(1,0))
#print(nn_xor(1,1))




#Task:4
#Printing different activation functions and find application
#All of the following activation functions has an advantage over perceptions as
#they are non-linear activation functions
#and can thus create non-linear decision boundries
#combining this with backprop allows for clasification of nonlinear 
#decision boundraries. Can also be used for non-linear regression.


# 100 linearly spaced numbers
x = np.linspace(-5,5,100)
%matplotlib inline



y1 =  np.exp(x)/(1+np.exp(x))
plt.plot(x,y1, 'r')
#Sigmoid function has issues with vanishing grads when values get big in  
#absolute value, a common choice for activation function in neural networks 
print("Sigmoid")
plt.show()  

plt.plot()
print("Sigmoid derivative")
plt.plot(x,y1*(1-y1))
plt.plot()
```

```
plt.plot(x,np.maximum(0,x), 'r')
print("Relu")
#More modern activation function, small issue with discontinuity if values are 
#identical to 0. Allows a neuron to fire more stongly to stronger inputs, but 
#risks overflow errors if implemented poorly. 
plt.show()  

print("Derivative of relu")
plt.plot(x,np.maximum(0,x/abs(x)))
plt.show()
```

```
plt.plot(x,np.tanh(x), 'r')
print("tanh")
#tanh is just a stretched version of sigmoid. Also one of the regular
#trig functions
plt.show()  

print("Derivative of tanh")
plt.plot(x,1-np.tanh(x)**2)
plt.show()
```

```
#Linear activation function, not used as frequently due to
#its inability to create non-linear decision boundraries
plt.plot(x,x, 'r')
print("linear")
plt.show()  

print("Derivative of linear")
plt.plot(x,x/np.abs(x),"r")
plt.show()
```

```
plt.plot(x,np.heaviside(x,1), 'r')
print("heaviside function, or hardlim")
#Can be used for binary clasification, can be used for the output layer of a nn
#
plt.show()  

print("Derivative is 0, and function is discountinous")
```
